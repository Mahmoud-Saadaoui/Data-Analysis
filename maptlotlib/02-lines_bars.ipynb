```python
import pandas as pd

df = pd.read_csv("car_malfunctions.csv")
 
df.info()
```

```python
import matplotlib.pyplot as plt

ypoints = df['malfunctions']
plt.plot(ypoints)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ls = '-')
plt.show()
```

```python
plt.plot(ypoints, color = 'red', linewidth = '10')
```
