DataFrame:
sert à stocker, manipuler et analyser des données tabulaires en Python, un peu comme une table SQL ou une feuille Excel.

```python
import pandas as pd
df = pd.DataFrame()
print (df)
```

```python
data = {
    "name": ["Ali", "Sami", "Nour"],
    "age": [22, 25, 21],
    "score": [15.5, 18, 14]
}
df = pd.DataFrame(data)
print (df)
```

```python
data = ["Ali", "Sami", "Nour"]
df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data, columns=["names"]
                  print(df)
```

```python
df = pd.DataFrame(data, columns=["names"])
print(df)
```
