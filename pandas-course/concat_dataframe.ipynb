```python
import pandas as pd

data_one = {
    "col_a": [1,2,3,4],
    "col_b": ["a", "b", "c", "d"],
    "col_c": [True, False, False, False]
}

data_two = {
    "col_a": [1,21,3,14],
    "col_b": ["m", "b", "n", "d"],
    "col_c": [True, False, True, True]
}

df_one = pd.DataFrame(data_one)
df_two = pd.DataFrame(data_two)

df_one
```

```python
df_two
```

```python
df = pd.concat([df_one, df_two])
df
```

```python
df = pd.concat([df_one, df_two], keys=["df1", "df2"])
df
```

```python
df = pd.concat([df_one, df_two], ignore_index=True)
df
```

```python
df = pd.concat([df_one, df_two], axis=1)
df
```
